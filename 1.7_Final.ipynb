```python
relationship_df= relationship_df.reset_index(
    
)
```

```python
# Create a graph from a pandas dataframe

G = nx.from_pandas_edgelist(relationship_df,
                                               source = "source",
                target = "target",
                edge_attr =  "value",
                create_using = nx.Graph())
```

```python
plt.figure(figsize = (10,10))
pos = nx.kamada_kawai_layout (G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
```

```python
from pyvis.network import Network
```

```python
# Define net object
net = Network()
net = Network(notebook = True,width="900px", height="800px", bgcolor='#222222', font_color='white')

# Define nodes from the G object
Node_degree = dict(G.degree)

# Setting up node size attributes
net = Network(notebook=True, width="900px", height="800px", bgcolor='#222222', font_color='white')
net.from_nx(G)
net.show_buttons(filter_= True)
net.repulsion()
net.show("relationship_df.html")
```

```python
import community as community_louvain
import networkx as nx

# Generate a graph and apply community detection
partition = community_louvain.best_partition(G)
nx.set_node_attributes(G, partition, 'community')
```

```python
from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k, v in new_com.items()}
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width="1300px", height="700px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_ = True)
com_net.repulsion()
com_net.show("20th_communities_leiden.html")
```

```python
import networkx as nx
```

```python
# Degree centrality

degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
```

```python
print("Degree Centrality:", degree_centrality)
print("Closeness Centrality:", closeness_centrality)
print("Betweenness Centrality:", betweenness_centrality)
```

```python
plt.figure(figsize=(10, 7))
nx.draw(G, with_labels=True, node_color=list(degree_centrality.values()), 
        cmap=plt.cm.plasma, node_size=[v * 1000 for v in degree_centrality.values()])
plt.title('Degree Centrality')
plt.show()
```

1. **Degree Centrality:**
   - Nodes with higher degree centrality are those with the most connections, indicating that they have direct influence over many nodes or in this case countries. 
we observe that the United States, Japan, and Germany display the most vibrant colors, indicating they have the highest degree of connectivity in the network. This suggests that these countries had substantial influence in global affairs during the 20th century. Notably, countries with lighter purple hues represent the next tier of connectivity and are key players in World War II. Germany, represented in orange, was a central Axis power, which aligns with its high connectivity in the dataset, reflecting its significant influence during that time.

The United States, the most vibrant node on the map, is a major player due to its extensive connections with other countries, particularly as a part of the Allied powers. Interestingly, Great Britain, while connected to Europe, is not as prominent in the color intensity, which may point to its more specialized connections.

   - **Color:** The color intensity for degree centrality visualizations is based on the degree of connectivity, with **darker colors** indicating **higher degree centrality**. These nodes are typically larger, representing more connections.

```python
plt.figure(figsize=(10, 7))
nx.draw(G, with_labels=True, node_color=list(betweenness_centrality.values()), 
        cmap=plt.cm.coolwarm, node_size=[v * 1000 for v in betweenness_centrality.values()])
plt.title('Betweenness Centrality')
plt.show()
```

2. **Betweenness Centrality:**
   - Nodes with high betweenness centrality serve as bridges between different parts of the network or systems or in this case countries, globally when analyizing the key words on the wikipedia page.
   - Notably, the United States appears as the largest and warmest node on the network, signifying its central role in global information flow and connectivity. This is evident through its involvement in major geopolitical and economic events during the 20th century, such as its interactions with Russia during the Cold War, its economic interests in the Middle East, and its foreign aid to countries like Libya and India during periods of political change. These examples highlight why the U.S. holds such a prominent position in the network.
   - **Color:** **Cooler colors (e.g., blue)** indicate **lower betweenness centrality**, while **warmer colors (e.g., red)** indicate **higher betweenness centrality**, highlighting the critical nodes that connect distinct parts of the network.

3. **Closeness Centrality:**
   Nodes with higher closeness centrality have shorter average paths to all other nodes, making them more efficient in communication across the network. We can see that the country identifiers on the map are hard to see since they're close together. This makes sense because it's showing the direct communication of networks between each country. For example, we can see on the outer portion of the map that Singapore and Thailand have a short distance between each other and also have a strong relationship with Japan for information. This might be because of their geographical proximity between the countries which affected that area in the 20th Century. 

**Color:** The **color intensity in closeness centrality** visualizations ranges from **light to dark**, where **darker nodes** indicate **higher closeness centrality**. These nodes are strategically positioned to minimize the distance between other nodes in the network.
