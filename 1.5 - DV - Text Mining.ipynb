```python
import spacy
from collections import Counter
import matplotlib.pyplot as plt
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
nouns = [token.text for token in doc if token.pos_ == "NOUN"]
verbs = [token.text for token in doc if token.pos_ == "VERB"]
adjectives = [token.text for token in doc if token.pos_ == "ADJ"]
```

```python
noun_counts = Counter(nouns)
verb_counts = Counter(verbs)
adj_counts = Counter(adjectives)
```

```python
noun_counts = Counter(nouns)
verb_counts = Counter(verbs)
adj_counts = Counter(adjectives)
```

```python
top_nouns = noun_counts.most_common(15)
top_verbs = verb_counts.most_common(15)
top_adjectives = adj_counts.most_common(15)
```

```python
top_nouns = noun_counts.most_common(15)
top_verbs = verb_counts.most_common(15)
top_adjectives = adj_counts.most_common(15)
```

```python
def plot_top_words(top_words, title, labels, color='skyblue'):
    words, counts = zip(*top_words)
    plt.figure(figsize=(10, 6))
    plt.bar(words, counts, color=color)
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
plot_top_words(top_nouns, 'Top 15 Nouns', nouns)
```

```python
plot_top_words(top_verbs, 'Top 15 Verbs', verbs, color='lightgreen')
```

```python
plot_top_words(top_adjectives, 'Top 15 Adjectives', adjectives, color='lightcoral')
```
